Distribution in details

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom, poisson, norm, expon, bernoulli
```

Normal: The distribution of height for 1000 adults, with an  average height of 170cm and a standard deviation of 10cm. The bell-shaped curve is characteristic of the normal distribution

```python
# Normal
Avg_height, Sd_height = 170, 10
# avg_height is the mean height while sd height is the standard deviation
heights_data = norm.rvs(Avg_height, Sd_height, size=1000)
heights_data
# rvs = random variates, norm = normal distribution(we want a graph that has a normal distribution)
```

```python
fig, axs = plt.subplots(1, 1, figsize=(10, 7))
sns.histplot(heights_data, kde=False, ax=axs,color="green")
axs.set_title("Normal")
axs.set_xlabel("Height")
axs.set_ylabel("Frequency")
```

```python
sns.displot(heights_data)
```

Exponential Distribution-Time specific events
The duration of 1000 customer service calls with an average call duration of 8 minutes. The distribution fits the exponential distribution.

```python
#Exponential
lambda_call = 1/8  # Average duration is 8 minutes  8/60
call_duration_data = expon.rvs(scale=1/lambda_call, size=1000)

fig, axs = plt.subplots(1, 1, figsize=(10, 7))
sns.histplot(call_duration_data, kde=False, ax=axs, color='orange')
axs.set_title('Exponential')
axs.set_xlabel('Call Duration')
axs.set_ylabel('Frequency')
```

```python
sns.displot(call_duration_data)
```

Poisson Distribution:The number of visitors to a website each hour over a 24hr period, with an average of 10 visitors per hour. The distribution of visitor counts per hour aligns with the poisson distribution

```python
lambda_visitors = 10  # Average number of visitors per hour
website_traffic_data = poisson.rvs(mu=10, size=24)  # 24 hours
# mu can also be equal to 10
fig, axs = plt.subplots(1, 1, figsize=(10, 7))
sns.histplot(website_traffic_data, kde=False, ax=axs, color='blue')
axs.set_title('Poisson')
axs.set_xlabel('Site Traffic')
axs.set_ylabel('Frequency')
```
