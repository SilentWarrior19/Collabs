# **Scatter plot with a linear trend line**


```python
import matplotlib.pyplot as plt
import numpy as np

# Simulated data - Square footage as independent variable (x) and house prices as dependent variable (y)
np.random.seed(42)
square_footage = np.random.randint(800, 3500, 50)  # Simulated square footage data
house_prices = 150000 + 200 * square_footage + np.random.normal(0, 10000, 50)  # Simulated house prices

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(square_footage, house_prices, label='House Prices')

# Fit a linear trend line (Linear Regression line)
m, b = np.polyfit(square_footage, house_prices, 1)  # Coefficients for the linear fit
plt.plot(square_footage, m * square_footage + b, color='red', label='Linear Trend Line')

plt.xlabel('Square Footage')
plt.ylabel('House Prices')
plt.title('House Prices vs Square Footage')
plt.legend()
plt.grid(True)
plt.show()
```

# **Multiple Linear Regression predicting house prices based on square footage, bedrooms, and location**

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from sklearn.linear_model import LinearRegression

# Simulated data - Square footage, bedrooms, location coordinates, and house prices
np.random.seed(42)
square_footage = np.random.randint(800, 3500, 100)
bedrooms = np.random.randint(1, 6, 100)
location = np.random.randint(1, 11, 100)
house_prices = 150000 + 200 * square_footage + 10000 * bedrooms - 3000 * location + np.random.normal(0, 10000, 100)

# Fit Multiple Linear Regression
X = np.column_stack((square_footage, bedrooms, location))
model = LinearRegression().fit(X, house_prices)

# Create a meshgrid for the predictors
x_surf, y_surf = np.meshgrid(np.linspace(square_footage.min(), square_footage.max(), 10),
                             np.linspace(bedrooms.min(), bedrooms.max(), 10))
x_surf, y_surf = x_surf.flatten(), y_surf.flatten()
z_surf = model.predict(np.array([x_surf, y_surf, np.full_like(x_surf, location.mean())]).T)
x_surf, y_surf = x_surf.reshape(-1, 10), y_surf.reshape(-1, 10)
z_surf = z_surf.reshape(10, 10)

# 3D Scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(square_footage, bedrooms, house_prices, c='blue', marker='o', label='Actual House Prices')
ax.plot_surface(x_surf, y_surf, z_surf, alpha=0.5, cmap='viridis', label='Regression Surface')

ax.set_xlabel('Square Footage')
ax.set_ylabel('Bedrooms')
ax.set_zlabel('House Prices')
ax.set_title('Multiple Linear Regression: House Prices Prediction')

plt.show()
```

#**Scatter Plot with Residuals:**

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Generating sample data
np.random.seed(42)
X = np.random.rand(100, 1) * 10
y = 2 * X + np.random.randn(100, 1) * 2  # Simulated linear relationship with noise

# Fit a Linear Regression model
model = LinearRegression()
model.fit(X, y)
predicted = model.predict(X)

# Calculating residuals
residuals = y - predicted

# Scatter plot of residuals against predicted values
plt.figure(figsize=(8, 6))
plt.scatter(predicted, residuals, alpha=0.8)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

# **Leverage vs. Residual Squared:**

```python
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

# Generating sample data
np.random.seed(42)
X = np.random.rand(100, 1) * 10
y = 2 * X + np.random.randn(100, 1) * 2  # Simulated linear relationship with noise

# Fit a Linear Regression model using statsmodels
X_with_const = sm.add_constant(X)
model = sm.OLS(y, X_with_const).fit()

# Calculating residuals and leverage
residuals = model.resid
leverage = model.get_influence().hat_matrix_diag

# Scatter plot of leverage vs. residuals squared
plt.figure(figsize=(8, 6))
plt.scatter(leverage, residuals**2, alpha=0.8)
plt.xlabel('Leverage')
plt.ylabel('Residuals Squared')
plt.title('Leverage vs. Residuals Squared')
plt.show()
```

# Histogram and a Q-Q plot to visualize the distribution of residuals from a regression model

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Generating sample data
np.random.seed(42)
X = np.random.rand(100, 1) * 10
y = 2 * X + np.random.randn(100, 1) * 2  # Simulated linear relationship with noise

# Fit a Linear Regression model
model = LinearRegression()
model.fit(X, y)
predicted = model.predict(X)

# Calculating residuals
residuals = y - predicted

# Plotting Histogram of Residuals
plt.figure(figsize=(8, 6))
plt.hist(residuals, bins=20, edgecolor='black')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.grid(True)
plt.show()

# Plotting Q-Q (Quantile-Quantile) Plot
plt.figure(figsize=(8, 6))
from scipy import stats
stats.probplot(residuals.flatten(), dist="norm", plot=plt)
plt.title('Q-Q Plot of Residuals')
plt.grid(True)
plt.show()
```

# Homoscedasticity (Equal Spread of Residuals)

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Generating sample data with homoscedasticity
np.random.seed(42)
X = np.random.rand(100, 1) * 10
y = 2 * X + np.random.randn(100, 1) * 2  # Simulated linear relationship with constant variance

# Fit a Linear Regression model
model = LinearRegression()
model.fit(X, y)
predicted = model.predict(X)

# Calculating residuals
residuals = y - predicted

# Scatter plot of residuals against predicted values for homoscedasticity
plt.figure(figsize=(8, 6))
plt.scatter(predicted, residuals, alpha=0.8)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.title('Homoscedasticity: Residual Plot')
plt.grid(True)
plt.show()
# Reliable and unbiased model
```

#  Heteroscedasticity (Unequal Spread of Residuals)


```python
# Generating sample data with heteroscedasticity
np.random.seed(42)
X = np.random.rand(100, 1) * 10
y = 2 * X + X**2 + np.random.randn(100, 1) * 2  # Simulated non-linear relationship causing varying variance

# Fit a Linear Regression model
model = LinearRegression()
model.fit(X, y)
predicted = model.predict(X)

# Calculating residuals
residuals = y - predicted

# Scatter plot of residuals against predicted values for heteroscedasticity
plt.figure(figsize=(8, 6))
plt.scatter(predicted, residuals, alpha=0.8)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.title('Heteroscedasticity: Residual Plot')
plt.grid(True)
plt.show()
# Unreliable and biased model
```
